# Import Data

# Application Train and Test

```python
import pandas as pd
```

```python
df_train = pd.read_csv("application_train.csv")
df_train.head()
```

```python
df_train.shape
```

```python
df_test = pd.read_csv("application_test.csv")
df_test.head()
```

```python
df_test.shape
```

```python
# Find Unique value from categorical features

for col in df_train.columns[df_train.dtypes == 'object']:
    c_train = set(df_train[col].unique())
    c_test = set(df_test[col].unique())
    diff = c_train - c_test
    if len(diff) > 0:
        print('feature ' + col + ' has unique values: ', diff)
```

```python
import numpy as np
```

```python
# Change the Unique value to Nan Value

df_train['CODE_GENDER'] = df_train['CODE_GENDER'].map(lambda x: x if x != 'XNA' else np.nan)
df_train['NAME_INCOME_TYPE'] = df_train['NAME_INCOME_TYPE'].map(lambda x: x if x != 'Maternity leave' else np.nan)
df_train['NAME_FAMILY_STATUS'] = df_train['NAME_FAMILY_STATUS'].map(lambda x: x if x != 'Unknown' else np.nan)
```

```python
# Create the label encoder on categorical features

from sklearn.preprocessing import LabelEncoder 
```

```python
# Label Encoder on data train

encoder = LabelEncoder()
for col in df_train.columns[df_train.dtypes=='object']:
    df_train[col] = encoder.fit_transform(df_train[col])
```

```python
df_train.head()
```

```python
# Label Encoder on Data test

for col in df_test.columns[df_test.dtypes=='object']:
    df_test[col] = encoder.fit_transform(df_test[col])
```

```python
df_test.head()
```

```python
bureau = pd.read_csv("bureau.csv")
bureau.head()
```

```python
bureau_balance = pd.read_csv("bureau_balance.csv")
bureau_balance.head()
```

```python
# One hot encoding on Bureau Balance

bureau_balance = pd.get_dummies(bureau_balance)
```

```python
bureau_balance.head()
```

```python
# Merge bureau balance on bureau

for col in bureau_balance.drop(columns=['SK_ID_BUREAU']).columns:
    res = bureau_balance.groupby(by = 'SK_ID_BUREAU')[col].mean()
    bureau['Balance_' + str(col)] = bureau['SK_ID_BUREAU'].map(lambda x: res[x] if x in res.index else np.nan)
```

# Previous Application

```python
prev_app = pd.read_csv("previous_application.csv")
prev_app.head()
```

```python
prev_app.shape
```

```python
# Labeling Categorical Features on Previous Application dataframe

for col in prev_app.columns[prev_app.dtypes == 'object']:
    if prev_app[col].nunique() == 2:
        d = prev_app[col].value_counts()
        prev_app[col] = prev_app[col].map({d[0]: 0, d[1]: 1})
    else:        
        for val in prev_app[col].unique():
            prev_app[str(col) + '_' + str(val)] = (prev_app[col] == val).map(int)
        prev_app.drop(columns=col, inplace = True)
```

```python
prev_app.shape
```

```python
# Merge Install Payments and Previous Application

for col in inst_pay.drop(columns=['SK_ID_PREV', 'SK_ID_CURR']).columns:
    res = inst_pay.groupby(by = 'SK_ID_PREV')[col].mean()
    prev_app['IP_' + col] = prev_app['SK_ID_PREV'].map(lambda x: res[x] if x in res.index else np.nan)
```

```python
prev_app.shape
```

```python
# Merge Credit Card Balance and Previous Application

for col in cc_balance.drop(columns=['SK_ID_PREV', 'SK_ID_CURR']).columns:
    res = cc_balance.groupby(by = 'SK_ID_PREV')[col].mean()
    prev_app['CCB_' + col] = prev_app['SK_ID_PREV'].map(lambda x: res[x] if x in res.index else np.nan)
```

```python
prev_app.shape
```

```python
# Merge Posh Cash Balance and Previous Application

for col in posh_cash.drop(columns=['SK_ID_PREV', 'SK_ID_CURR']).columns:
    res = posh_cash.groupby(by = 'SK_ID_PREV')[col].mean()
    prev_app['POS_' + col] = prev_app['SK_ID_PREV'].map(lambda x: res[x] if x in res.index else np.nan)
```

```python
prev_app.shape
```

```python
# define the new data frame

bureau_merged = bureau.copy()
prev_app_merged = prev_app.copy()
```

```python
bureau_merged.shape
```

```python
prev_app_merged.shape
```

# Merge All dataframe

```python
# Data train/test and Bureau Merged

for col in bureau_merged.drop(columns = ['SK_ID_CURR', 'SK_ID_BUREAU']).columns:
    res = bureau_merged.groupby(by = 'SK_ID_CURR')[col].mean()
    df_train['Bureau_' + str(col)] = bureau_merged['SK_ID_CURR'].map(lambda x: res[x] if x in res.index else np.nan)
    df_test['Bureau_' + str(col)] = bureau_merged['SK_ID_CURR'].map(lambda x: res[x] if x in res.index else np.nan)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
# Data train/test and Prev app merged

for col in prev_app_merged.drop(columns=['SK_ID_CURR', 'SK_ID_PREV']).columns:
    res = prev_app_merged.groupby(by = 'SK_ID_CURR')[col].mean()
    df_train['Prev_' + str(col)] = prev_app_merged['SK_ID_CURR'].map(lambda x: res[x] if x in res.index else np.nan)
    df_test['Prev_' + str(col)] = prev_app_merged['SK_ID_CURR'].map(lambda x: res[x] if x in res.index else np.nan)
```

```python
print(df_train.shape)
print(df_test.shape)
```
